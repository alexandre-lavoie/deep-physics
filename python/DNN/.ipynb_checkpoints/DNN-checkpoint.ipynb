# Classes

```python
class Particle:
    def __str__(self):
        return "Position: %s, Velocity: %s, Accleration: %s" % (self.position, self.velocity, self.acceleration)
    
    def __repr__(self):
        return "Position: %s, Velocity: %s, Accleration: %s" % (self.position, self.velocity, self.acceleration)
    
    def __init__(self, position=None, velocity=None, acceleration=None):
        if position = None:
            self.position = np.array([np.random.sample()*2-1,np.random.sample()*2-1]) # Position X, Y
        else:
            self.position = position
        
        if velocity = None:
            self.velocity = np.array([np.random.sample()*2-1,np.random.sample()*2-1]) # Velocity X, Y
        else:
            self.velocity = velocity
            
        if acceleration = None:
            self.acceleration = np.array([np.random.sample()*2-1,np.random.sample()*2-1]) # Acceleration X, Y
        else:
            self.acceleration = acceleration
        
        
    def apply_physics(self,dt):
        nextParticle = copy.deepcopy(self) # Copy to retain initial values
        nextParticle.position += self.velocity * dt
        for p in nextParticle.position: # Caps position to [-1,1]
            p = max(min(1,p),-1)
        nextParticle.velocity += self.acceleration * dt
        for v in nextParticle.velocity: # Caps velocity to [-1,1]
            v = max(min(1,v),-1)
        return nextParticle
    
    def predict(model):
        pred = model.predict(self.as_list())
        
    
    def as_list(self):
        return [self.position[0],self.position[1],self.velocity[0], self.velocity[1], self.acceleration[0], self.acceleration[1]]
    
    def physics_as_list(self,dt):
        n = self.apply_physics(dt)
        return [self.position[0],self.position[1],self.velocity[0], self.velocity[1], self.acceleration[0], 
                self.acceleration[1], n.position[0], n.position[1], n.velocity[0], n.velocity[1]]
    
    def get_actual_trajectory(self, dt, time):
        temp = copy.deepcopy(self)
        position = []
        velocity = []
        for _ in range(time):
            position.append(temp.position)
            velocity.append(temp.velocity)
            temp = temp.apply_physics(dt)
        return position, velocity
    
    
```

```python
class DNN:
    def __init__(self, input_size, output_size, dropout=0.4, model=None):
        if model == None:
            self.model = get_model
        else:
            self.model = model
    
    def get_model(self):
        optimizer = Adam()
        input_layer = Input(shape=(self.input_size,), name="Input") 
        dense_layer = Dense(64, activation='relu')(input_layer)
        dense_layer = Dropout(self.dropout)(dense_layer)
        dense_layer = Dense(128, activation='relu')(dense_layer)
        dense_layer = Dropout(self.dropout)(dense_layer)
        dense_layer = Dense(64, activation='relu')(dense_layer)
        dense_layer = Dropout(self.dropout)(dense_layer)
        output_layer = Dense(self.output_size, activation='tanh', name="Output")
        model = Model(input_layer, output_layer)
        model.compile(loss="mean_squared_error", optimizer=optimizer)
        return model
    
    def train(data, labels, epochs = 150, batch_size = 32, save = True):
        self.model.fit(data, labels, epochs=epochs, batch_size=batch_size, verbose=1)
        if save: self.model.save("Model.h5")
        
    def predict(particle):
        return self.model.predict(np.array(particle).reshape(-1,6))
```

```python
def get_samples(dt):
    object_list = []
    step = []
    for _ in range(number_of_particle):
        obj = Particle()
        object_list.append(obj.as_list())
        step.append(obj.physics_as_list(dt))
    return np.array(obj), np.array(step)
```

```python
X, y = get_samples(0.1)
```

```python
model = DNN(input_size=input_size, output_size=output_size)
model.train(X,y,epochs=epochs,batch_size=batch_size)
```
